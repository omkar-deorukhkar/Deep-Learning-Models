```
import pandas as pd

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data')


names = ['sno','clump_th','csz_uni','csh_uni','mar_adh','secs','bare_n','bc','nn','mito','class']
df.columns = names
df = df.drop(columns = ['sno'])
df = df.replace({'?':np.nan})


```

```
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder



df = np.asarray(df)

df = df.astype('float32')


imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
df = imputer.fit_transform(df)

Y = df[:,-1:]
X = df[:,:-1]

scaler = MinMaxScaler(feature_range =(0,1))
X = scaler.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
Y_to = Y_test
OHE = OneHotEncoder()
Y = OHE.fit_transform(Y)

encoded = {0:'benign',1:'malignant',2:'benign', 4:'malignant'}


```

```
from keras.models import Sequential
from keras.layers import Dense, Activation

model = Sequential()
model.add(Dense(81, input_dim = X_train.shape[1], activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(2, activation='softmax'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train,Y_train,batch_size=25,epochs=100,shuffle=False)
```
